Проверка перемножений:

```python
import numpy as np

sizes = [100, 250, 500, 750, 1000]

for size in sizes:
    matrix_A = np.loadtxt(f'data_and_res/matrix_A_{size}.txt')
    matrix_B = np.loadtxt(f'data_and_res/matrix_B_{size}.txt')
    result_AxB = np.loadtxt(f'data_and_res/res_AxB_{size}.txt')

    result_python = np.dot(matrix_A, matrix_B)

    if np.array_equal(result_python, result_AxB):
        print(f'Результаты для размера {size} совпадают!')
    else:
        print(f'Результаты для размера {size} не совпадают!')
```

Графическое представление зависимости среднего времени перемножения от размеров матриц

```python
import matplotlib.pyplot as plt

sizes = []
times = []
with open('data_and_res/average_times.txt', 'r') as file:
    for line in file:
        size, time = map(int, line.strip().split('-'))
        sizes.append(size)
        times.append(time)


plt.plot(sizes, times, marker='o', linestyle='-')
plt.title('Среднее время')
plt.xlabel('Размерм матриц')
plt.ylabel('Время (мс)')
plt.grid(True)
plt.show()
```
